```python
import numpy as np
from shape import Circle, Triangle
```

```python
def gen_dataset(num_shapes,num_points):
    shapes = []
    for _ in range(num_shapes):
        shape = Circle(radius=1,n_points=num_points)
        shape = Triangle( n_points=num_points)
        shapes.append(shape)
    return shapes
```

```python
shapes = gen_dataset(50,100)
for shape in shapes:
    shape.plot()
```
